# Load necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#resize your graph
plt.figure(figsize=(5, 3), dpi=200)

#line1

#keyword argument notation
# plt.plot(x, y, label='2x', color='red', linewidth='2', marker='.', linestyle='--', markersize='10', markeredgecolor='blue')

#use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x, y,  'r.--', label='2x')

##line2
#select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)
print(x2)

#plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'y', label='X^2')

#plot remainder graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'y--')


#add a title(specify font parameter)
plt.title('My First Graph', fontdict={'fontname': 'Monospace', 'fontsize': 20})

#add X Y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

#add X, Y tickmarks(scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])


plt.legend()


#save graph
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

```python
#ex:
x = [2, 4, 6, 8]
y = [10, 12, 14, 16]
plt.plot(x, y, 'r', label='2x')
plt.legend()
plt.show()
```

## # Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1 , 5, 3]
bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')


plt.figure(figsize=(6, 3))
plt.show()
```
